## STEP 1: Input loading
### 1.1 JSON import
### 1.2 Transcriptome metadata import
### 1.3 Create final dataframe JSON and CSV information

```python
def formatID(x):
    return x.replace('01','')

def repno(x):
    try:
        d = int(x)
    except:
        d = np.nan
    return d

def remzero(x):
    if x == 0:
        return np.nan
    else:
        return x
    
def annotyesno(x):
    if x == 'Yes':
        return 1
    else:
        return 0
    
def contigcountno(x):
    if x  == 'TSA':
        return 0
```

```python
def load_JSON(JSON_FILE):
    ### JSON import
    with open(JSON_FILE, encoding='utf-8') as data_file:
        JSON_DATA = json.loads(data_file.read())
    return JSON_DATA


def load_TSA_metadata(TSA_FILE):
    ### CSV import
    TSA_CSV = open(TSA_FILE)

    allLines = [x for x in TSA_CSV.readlines()]
    columns = allLines[0].strip('\n').split('\t')
    rows = []
    for line in allLines[1:] :
        line = line.strip('\n')
        line = line.split('\t')
        rows.append(line)

    TSA_METADATA = pd.DataFrame(data=rows,columns=columns)
    TSA_METADATA['id'] = TSA_METADATA['id'].apply(formatID)
    return TSA_METADATA


def build_TSA_dataframe(JSON_DATA, TSA_METADATA):
    '''
    Read JSON object
    Add results of oskar analysis to TSA dataframe 
    Compute specific variables for transcriptomes statistics
    '''
    already = []
    tsa_oskar = []
    for ID in JSON_DATA :
        if 'Transcriptome_analysis' in JSON_DATA[ID] :
            for analysis in JSON_DATA[ID]['Transcriptome_analysis']:
                try:
                    if len(analysis['result']['oskar']['validated']) != 0 :
                        tsa_oskar.append([analysis['tsa_id'], 'found'])
                    else :
                        tsa_oskar.append([analysis['tsa_id'], 'not found'])
                except:
                    print('No transcriptome analysis for this sequence: {} (identified manually)'.format(analysis['genome_id']))
                        
    TSA_oskar_results = pd.DataFrame(tsa_oskar, columns = ['id', 'oskar'])
    TSA_dataframe = TSA_oskar_results.merge(TSA_METADATA, on='id')
    TSA_dataframe['contigs_total_length_l'] = TSA_dataframe['contigs_total_length_l'].apply(repno)
    TSA_dataframe['contigs_count_l'] = [int(x) for x in TSA_dataframe['contigs_count_l']]
    TSA_dataframe['contigs_proteins_count_l'] = [int(x) for x in TSA_dataframe['contigs_proteins_count_l']]
    TSA_dataframe['notzeroprotcount'] = TSA_dataframe['contigs_proteins_count_l'].apply(remzero)
    TSA_dataframe['n50'] = TSA_dataframe['contigs_count_l'] / TSA_dataframe['contigs_total_length_l']
    TSA_dataframe['contigs_min_length_l'] = [int(x) for x in TSA_dataframe['contigs_min_length_l']]
    TSA_dataframe['contigs_max_length_l'] = [int(x) for x in TSA_dataframe['contigs_max_length_l']]
    TSA_dataframe['annotnumber'] = TSA_dataframe['contigs_annotated_s'].apply(annotyesno)
    return TSA_dataframe

def build_TSA_osk_table(JSON_FILE, TSA_FILE):
    '''
    Build up function to save the final TSA table into a CSV file
    '''
    JSON_DATA = load_JSON(JSON_FILE)
    TSA_METADATA = load_TSA_metadata(TSA_FILE)
    TSA_DF = build_TSA_dataframe(JSON_DATA, TSA_METADATA)
    TSA_DF.to_csv(os.path.join(data_path,'2.1/TSA_OSK_TABLE.csv'), index=False)

def build_genome_dataframe(JSON_DATA):
    '''
    Read JSON object
    Add results of oskar analysis to genome dataframe 
    Collect specific variables for genome statistics from NCBI database
    Save the final table into a CSV file
    '''
    genome_oskar = []
    already = []
    bar = progressbar.ProgressBar()
    for ID in bar(JSON_DATA) :
        if 'Genome_analysis' in JSON_DATA[ID] :
            for analysis in JSON_DATA[ID]['Genome_analysis']:
                if analysis['genome_id'] not in already :
                        if 'GC' in analysis['genome_id']:
                            already.append(analysis['genome_id'])
                            time.sleep(0.5)
                            genome_search = Entrez.esearch(db="assembly", term=analysis['genome_id'])
                            orgn_id = Entrez.read(genome_search)['IdList']
                            assembly_handle = Entrez.esummary(db="assembly", id=orgn_id)
                            summary = Entrez.read(assembly_handle, validate=False)['DocumentSummarySet']['DocumentSummary']

                            for tag in summary :

                                alt_loci_count = re.findall(r'<Stat category="alt_loci_count" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                chromosome_count = re.findall(r'<Stat category="chromosome_count" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                contig_count = re.findall(r'<Stat category="contig_count" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                contig_l50 = re.findall(r'<Stat category="contig_l50" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                contig_n50 = re.findall(r'<Stat category="contig_n50" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                non_chromosome_replicon_count = re.findall(r'<Stat category="non_chromosome_replicon_count" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                scaffold_count = re.findall(r'<Stat category="scaffold_count" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                scaffold_l50 = re.findall(r'<Stat category="scaffold_l50" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                scaffold_n50 = re.findall(r'<Stat category="scaffold_n50" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                total_length = re.findall(r'<Stat category="total_length" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]
                                ungapped_length = re.findall(r'<Stat category="ungapped_length" sequence_tag="all">([0-9]+)<\/Stat>', tag['Meta'])[0]

                            if len(analysis['result']['oskar']['validated']) != 0 : 
                                genome_oskar.append([analysis['genome_id'], 'found', int(alt_loci_count), int(chromosome_count), int(contig_count), int(contig_l50),  int(contig_n50), int(non_chromosome_replicon_count), int(scaffold_count), int(scaffold_l50), int(scaffold_n50), int(total_length), int(ungapped_length)])

                            else :
                                genome_oskar.append([analysis['genome_id'], 'not found', int(alt_loci_count), int(chromosome_count), int(contig_count), int(contig_l50), int(contig_n50), int(non_chromosome_replicon_count), int(scaffold_count), int(scaffold_l50), int(scaffold_n50), int(total_length), int(ungapped_length)])
                                
    GENOME_dataframe = pd.DataFrame(genome_oskar, columns = ['id', 'oskar', 'alt_loci_count', 'chromosome_count', 'contig_count', 'contig_l50', 'contig_n50','non_chromosome_replicon_count', 'scaffold_count', 'scaffold_l50', 'scaffold_n50', 'total_length', 'ungapped_length'])
    GENOME_dataframe['contigNumberperLength'] = GENOME_dataframe['contig_count'] / GENOME_dataframe['total_length']
    GENOME_dataframe['scaffoldNumberperLength'] = GENOME_dataframe['scaffold_count'] / GENOME_dataframe['total_length']
    return GENOME_dataframe

    
def build_genome_osk_table(JSON_FILE):
    '''
    Build up function to save the final genome table into a CSV file
    '''
    JSON_DATA = load_JSON(JSON_FILE)
    GENOME_DF = build_genome_dataframe(JSON_DATA)
    GENOME_DF.to_csv(os.path.join(data_path, '2.1/GENOME_OSK_TABLE.csv'), index=False)
```
