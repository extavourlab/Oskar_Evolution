# Hymenopteran family duplication
This script is used to plot 

```python
import ete3
import pandas as pd

from ete3 import Tree, faces, TreeStyle, TextFace
```

## Defining the isoforms, duplicated sequences, and names for each sequences

```python
df = pd.read_table('../Data/02_Oskar_analyses/2.7/nametable.tsv')
isoforms = [
    'GAUP02017617.1',
    'GFAF01022698.1',
    'GBLL01020427.1',
    'GBOS01032454.1',
    'XP_011057668.1',
    'XP_011057669.1',
    'XP_011635164.1',
    'XP_011635158.1',
    'GDFS01087539.1',
    'GDFS01087542.1'
    
]
duplicated= [
    'GBVD01033707.1',
    'GBVD01023175.1',
    'GAIW01011551.1',
    'GAIW01011550.1',
    'GAIW01009539.1',
    'GAJC01011221.1',
    'GAJC01009625.1',
    'GAXY02017799.1',
    'GAXY02013477.1',
    'GAYV02023556.1',
    'GAKJ01009501.1',
    'GBRL01010261.1',
    'GAKJ01010751.1',
    'GATC02018205.1',
    'GATC02016035.1',
    'GBLN01017162.1',
    'GBLN01013985.1',
    'XP_011175260.1',
    'XP_011165902.1',
    'GBWU01001740.1',
    'GBWU01000527.1',
]
```

```python
def make_tree(faces=False):
    T = ete3.Tree('../Data/02_Oskar_analyses/2.7/Hymenoptera_OSKAR_alignment_trimmed_0.5.nex.con.tre.newick')  
    for leaf in T:
        if 'GBKM01001402.1' in leaf.name:
            out1 = leaf
        if 'GBQG01014503.1' in leaf.name:
            out2 = leaf
    root = T.get_common_ancestor(out1, out2)
    T.set_outgroup(root)
    for leaf in T:
        n = leaf.name
        n = n.replace("'","")
        if faces:
            leaf.name = '  ' + n
            sp_name = df[df['ID'] == n]['Specie'].values[0].replace('_', ' ')
            sp_name = sp_name[0].upper() + sp_name[1:]
            fam_name = df[df['ID'] == n]['Family'].values[0]
            F = TextFace(sp_name, fstyle='italic', fsize=14)
            leaf.add_face(F, column=1, position="aligned")
            F = TextFace(fam_name, fsize=14)
            leaf.add_face(F, column=2, position="aligned")
        else:
            sp_name = df[df['ID'] == n]['Specie'].values[0].replace('_', ' ')
            sp_name = sp_name[0].upper() + sp_name[1:]
            fam_name = df[df['ID'] == n]['Family'].values[0]
            leaf.name = '  ' + fam_name + '  '
            F = TextFace(sp_name, fstyle='italic', fsize=16)
            F.margin_left = 5
            leaf.add_face(F, column=1)
        nstyle = ete3.NodeStyle()
        if n in duplicated:
            nstyle["fgcolor"] = "#b96565"
            nstyle["size"] = 22
        elif n in isoforms:
            nstyle["fgcolor"] = "#7c7195"
            nstyle["size"] = 22
        else:
            nstyle["fgcolor"] = "#8f8f8f"
            nstyle["size"] = 22
        leaf.set_style(nstyle)
    return T
```

## Defining the Tree Style for plotting

```python
ts = ete3.TreeStyle()
ts.scale = 180
ts.tree_width = 600
ts.draw_guiding_lines = True
ts.optimal_scale_level = 'mid'
ts.show_branch_support = False
ts.show_scale = False
ts.show_leaf_name = True
ts.complete_branch_lines_when_necessary = True
```

## Rendering linear version of the tree

```python
T = make_tree(faces=True)
ts.mode = "r"
T.render('%%inline', tree_style=ts)
```

```python
T.render("../Figures/SVG/F3_v1.svg", tree_style=ts)
```

## Rendering circular version of the tree

```python
ts.mode = "c"
T = make_tree(faces=False)
T.render('%%inline', tree_style=ts)
```

```python
T.render("../Figures/SVG/F3_v2.svg", tree_style=ts)
```
